```python
from lxml import html
import requests
import sys
import re

class Tester(object):

    def __init__(self, name):
        self.name = str(name).upper().strip()
        
        try:
            self.page = requests.get("https://finance.yahoo.com/quote/" + name + "/history?p=" + name)
            self.tree = html.fromstring(self.page.content)
            self.page_text = self.page.text
            
            info_start = self.page_text.find('HistoricalPriceStore')
            info_end = self.page_text.find(',"isPending"')
           
            self.info = self.page_text[info_start:info_end]   
           
        except Exception as e:
            print("ERROR:\n" + str(e.args) + "\n\nTerminating execution...")
            sys.exit()

    def get_table(self):
        
        day = self.info.split("},{")
        day_new = [days.replace('HistoricalPriceStore":{"prices":[{',"") for days in day]
        day1 =  [y.replace("}]","") for y in day_new]
        day2 = [y.replace('date":','') for y in day1]
        day3 = [y.replace('"open":','') for y in day2]
        day4 = [y.replace('"high":','') for y in day3]
        day5 = [y.replace('"low":','') for y in day4]
        day6 = [y.replace('"close":','') for y in day5]
        day7 = [y.replace('"volume":','') for y in day6]
        day8 = [y.replace('"adjclose":','') for y in day7]
        
        titles = ['date','open','high','low','close','volume','adjvolume']
        day_list = [x.split("','") for x in day8]
        day_list = [titles] + day_list
        return day_list
    
    
share=Tester('aapl')
share.get_table()
```
